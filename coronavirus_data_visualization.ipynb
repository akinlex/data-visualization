```python
names=['Country','Confirmed', 'Suspected', 'Recovered', 'Deaths']
cleaned_data = data_frame[names]
```

```python
cleaned_data.dropna(how='all')
```

```python
import seaborn as sns
```

```python
# Scatter plot visualization
# By default, relplot is scatter plot
sns.relplot(x='Recovered', y='Confirmed', hue='Country', palette="ch:r=-.8,l=.80", data=cleaned_data)
```

```python
# Lineplot visualization
# Country column is dropped to visualize only columns with figures
sns.lineplot(data=cleaned_data.drop(['Country'], axis=1))
```

```python
# Historam visualization
# kde = kernel density estimate
sns.distplot(cleaned_data['Recovered'], bins=10, kde=False)
```

```python
# Distribution plot
sns.distplot(cleaned_data['Recovered'], bins=10, kde=True)
```

```python
# Bar-chart can be created using the sns.countplot method 
sns.countplot(cleaned_data['Confirmed'])
```

```python
import matplotlib.pyplot as plt
```

```python
# Read new csv data-file
winemagdf = pd.read_csv('winemag_data.csv')
winemagdf
```

```python
winemagdf.describe()
```

```python
cleaned_wine_review_data= winemagdf.dropna(subset=['price'], axis=0)
cleaned_wine_review_data
```

```python
cleaned_wine_review_data.dtypes
```

```python
# Histogram using matplotlib

# If we pass it categorical data 
# like the points column from the wine-review dataset 
# it will automatically calculate how often each class occurs.

# Create figure and axis
fig, ax = plt.subplots()
# Plot histogram
ax.hist(cleaned_wine_review_data['points'])
# Set labels
ax.set_title('Wine Review Scores')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

```python
# Bar chart using matplotlib

# bar-chart isnâ€™t automatically calculating 
# the frequency of a category so we are going 
# to use pandas value_counts function to do this

# Create figure and axis
fig, ax = plt.subplots()
# Count number of frequency(number of occurrence)
point_count = cleaned_wine_review_data['points'].value_counts()
# set x and y axis
points_x = point_count.index 
frequency_y = point_count.values
# Create bar chart
ax.bar(points_x, frequency_y) 
# Set title and labels 
ax.set_title('Wine Review Scores') 
ax.set_xlabel('Points') 
ax.set_ylabel('Frequency')
```

```python
# Box plots
goodwines_lessthan_1000 = cleaned_wine_review_data[(cleaned_wine_review_data['points']>=90) & (cleaned_wine_review_data['price']<1000)]
sns.boxplot('points', 'price', data=goodwines_lessthan_1000)
```
